```python
#1 Loading Data

import pandas as pd
df = pd.read_csv("WBD_DATA.zip")
df
```

```python
# --- 2. Initial Data Exploration ---
# Understanding data structure, data types, and missing values.

print("\nDataFrame Information (data types, non-null counts):")
df.info()

print("\nDescriptive Statistics:")
print(df.describe())

print("\nNumber of missing values per column:")
print(df.isnull().sum())

# Check column names for consistency
print("\nColumn Names:")
print(df.columns)
```

```python
# --- 3. Data Cleaning ---

# a. Handling Missing Values
# Options:
# 1. Drop rows/columns with missing values (if few)
#    df.dropna(inplace=True) # Drops rows with any missing values
# 2. Impute missing values
#    For example, fill missing inflation values with the country's average inflation or global average.
#    For this example, we will fill missing numerical values with the median,
#    and categorical values with the mode.
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        if df[col].isnull().sum() > 0:
            median_val = df[col].median()
            df[col].fillna(median_val, inplace=True)
            print(f"Filled missing values in column '{col}' with median: {median_val}")
    elif df[col].dtype == 'object':
        if df[col].isnull().sum() > 0:
            mode_val = df[col].mode()[0]
            df[col].fillna(mode_val, inplace=True)
            print(f"Filled missing values in column '{col}' with mode: {mode_val}")

print("\nNumber of missing values after cleaning:")
print(df.isnull().sum())

# b. Handling Duplicates
# Remove duplicate rows if any
initial_rows = df.shape[0]
df.drop_duplicates(inplace=True)
if df.shape[0] < initial_rows:
    print(f"\nRemoved {initial_rows - df.shape[0]} duplicate rows.")
else:
    print("\nNo duplicate rows found.")

# c. Data Type Correction
# Ensure 'Year' column is integer and inflation columns are float.
# Adjust 'Year' if the column name is different, e.g., 'Time' or 'Date'
if 'Year' in df.columns:
    df['Year'] = pd.to_numeric(df['Year'], errors='coerce').astype('Int64') # Using Int64 to handle NaN
    print("\n'Year' column converted to integer type.")

# Identify inflation columns. Assuming there are columns like 'Inflation_Rate' or 'Annual_Inflation'
# You need to adjust this based on the column names in your dataset.
inflation_cols = [col for col in df.columns if 'inflation' in col.lower()]
if inflation_cols:
    for col in inflation_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        # Fill NaNs that might appear from 'coerce' (non-numeric values)
        if df[col].isnull().sum() > 0:
            df[col].fillna(df[col].median(), inplace=True) # Fill with median
            print(f"Column '{col}' converted to float type and NaNs filled.")
else:
    print("Warning: No columns identified as 'inflation'. Please check your column names.")
```

```python
# --- 4. Data Transformation & Feature Engineering ---

# a. Filter Data for Year Range 2010-2025
# Ensure 'year' column exists and has been cleaned
if 'year' in df.columns:
     df_filtered = df[(df['year'] >= 2010) & (df['year'] <= 2025)].copy()
     print(f"\nData filtered for years 2010-2025. Number of rows: {df_filtered.shape[0]}")
else:
     df_filtered = df.copy()
     print("\n'year' column not found, cannot filter by year.")


# b. Calculate Global Average Inflation Per Year (for trends)
# Assuming an 'Inflation_Rate' column has been identified or you adjust it
if inflation_cols:
    # Using the first identified inflation column as an example
    main_inflation_col = inflation_cols[0]
    global_inflation_trend = df_filtered.groupby('year')[main_inflation_col].mean().reset_index()
    global_inflation_trend.rename(columns={main_inflation_col: 'Global_Average_Inflation'}, inplace=True)
    print("\nGlobal Average Inflation per Year calculated:")
    display(global_inflation_trend.head())
else:
    print("Cannot calculate global average inflation as main inflation column is not identified.")

# c. Feature Engineering for "Multi-Indicator Classification Approach"
# This is the part where you will integrate results from IBM Granite.
# If IBM Granite provides classifications (e.g., 'Stable', 'Watchlist', 'High Risk')
# based on multiple indicators (inflation, GDP, unemployment, etc.),
# you will add this column to your DataFrame.

# Placeholder example: Assume you have logic or output from IBM Granite
# that classifies each row (country-year) into a stability category.
# You can use simple rules as an example or integrate API output.

# Example of simple classification based on inflation rate (THIS IS JUST AN EXAMPLE, IBM GRANITE WILL BE MORE COMPLEX)
if main_inflation_col in df_filtered.columns:
    def classify_stability(inflation_rate):
        if inflation_rate <= 3e14: # Adjusted threshold based on observed data
            return 'Stable'
        elif 3e14 < inflation_rate <= 7e14: # Adjusted threshold based on observed data
            return 'Watchlist'
        else:
            return 'High Risk'

    df_filtered['Economic_Stability_Class'] = df_filtered[main_inflation_col].apply(classify_stability)
    print("\n'Economic_Stability_Class' column added (simple classification example):")
    display(df_filtered[['country_name', 'year', main_inflation_col, 'Economic_Stability_Class']].head())
else:
    print("\nCannot add 'Economic_Stability_Class' column as main inflation column is missing.")

# If you have other indicators (GDP, Unemployment, etc.) in your dataset,
# you can combine them and use them for more sophisticated classification.
# For example, if you have 'GDP_Growth' and 'Unemployment_Rate' columns:
# df_filtered['Composite_Score'] = (df_filtered[main_inflation_col] * 0.5) + (df_filtered['GDP_Growth'] * -0.3) + ...
# Then classify based on 'Composite_Score' or use an AI model.
```

```python
# --- 5. Data Aggregation for Dashboard ---
# Prepare data in a format that is easiest to visualize in Excel.

# a. Data for Global Inflation Trend (already calculated in step 4b)
# global_inflation_trend

# b. Data for Inflation per Country (Average inflation per country over the 2010-2025 period)
if main_inflation_col in df_filtered.columns:
    inflation_per_country = df_filtered.groupby('country_name')[main_inflation_col].mean().reset_index()
    inflation_per_country.rename(columns={main_inflation_col: 'Average_Inflation_2010_2025'}, inplace=True)
    print("\nAverage Inflation per Country (2010-2025):")
    print(inflation_per_country.head())

# c. Data for Inflation by Stability Classification
if 'Economic_Stability_Class' in df_filtered.columns:
    inflation_by_class = df_filtered.groupby('Economic_Stability_Class')[main_inflation_col].mean().reset_index()
    inflation_by_class.rename(columns={main_inflation_col: 'Average_Inflation'}, inplace=True)
    print("\nAverage Inflation by Stability Classification:")
    print(inflation_by_class)

# d. Combined/Final Data for Dashboard
# For an Excel dashboard, it's often easier to have one large, clean, and complete table
# that can be used to create PivotTables and charts.
# Ensure df_filtered has all the columns you need (Country, Year, Inflation, Classification, etc.)
df_final_dashboard = df_filtered[['country_name', 'year', main_inflation_col, 'Economic_Stability_Class']].copy()
# Add other relevant columns if available (e.g., GDP, Unemployment)
# df_final_dashboard = df_filtered[['country_name', 'year', main_inflation_col, 'Economic_Stability_Class', 'GDP_Growth', 'Unemployment_Rate']].copy()

print("\nFinal data ready for dashboard (first few columns):")
print(df_final_dashboard.head())
print(f"Number of rows and columns in final data: {df_final_dashboard.shape}")
```
